# key_words

```python
import collections
from wordcloud import WordCloud, STOPWORDS

def Key_Words(list_of_text, n):
    s = ' '.join(list_of_text)
    s=s.lower().split()
    counter = collections.Counter(s)
    counter= sorted(counter.items(), key=lambda x:x[1],reverse=True)
    sum=0
    temp=[]
    for i in counter:
        if(i[0] not in STOPWORDS and i[0].isalpha() and len(i[0])>2):
            sum+=1
            temp.append(i)
        if(sum==n):
            break
    temp=dict(temp)
    array = [ {'key' : i, 'value' : temp[i]/len(s)} for i in temp]
    return array
```

```python
import collections
from wordcloud import WordCloud, STOPWORDS
import itertools
def Key_Words(list_of_text, n):
    a=[list(set(q.lower().split())) for q in list_of_text]
    flat_list = list(itertools.chain(*a))
    counter = collections.Counter(flat_list)
    counter= sorted(counter.items(), key=lambda x:x[1],reverse=True)
    sum=0
    temp=[]
    for i in counter:
        if(i[0] not in STOPWORDS and i[0].isalpha() and len(i[0])>2):
            sum+=1
            temp.append(i)
        if(sum==n):
            break
    temp=dict(temp)
    array = [ {'key' : i, 'value' : temp[i]/len(list_of_text)} for i in temp]
    return array
```

```python
def top_n(n, query,input_emb,text,model):
    query = [query]
    query_emb = model.encode(query, convert_to_tensor=True)
    print('---------begin----------')
    time1=time()
    cosine_scores = util.cos_sim(input_emb, query_emb)
    cosine_scores = cosine_scores.reshape(1,-1)[0]
    answer = cosine_scores.argsort(descending=True).cpu()
    for item in answer[:n]:
        print(cosine_scores[int(item)], text[int(item)%300000])
    print('---------end------------')
    print(time()-time1)
    temp=[]
    for item in answer[:n]:
        temp.append(text[int(item)])
    return(temp)
```
