# key_words

```python
import collections
from wordcloud import WordCloud, STOPWORDS

def Key_Words(list_of_text, n):
    s = ' '.join(list_of_text)
    s=s.lower().split()
    counter = collections.Counter(s)
    counter= sorted(counter.items(), key=lambda x:x[1],reverse=True)
    sum=0
    temp=[]
    for i in counter:
        if(i[0] not in STOPWORDS and i[0].isalpha() and len(i[0])>2):
            sum+=1
            temp.append(i)
        if(sum==n):
            break
    temp=dict(temp)
    array = [ {'key' : i, 'value' : temp[i]/len(s)} for i in temp]
    return array
```

```python
import collections
from wordcloud import WordCloud, STOPWORDS
import itertools
def Key_Words(list_of_text, n):
    a=[list(set(q.lower().split())) for q in list_of_text]
    flat_list = list(itertools.chain(*a))
    counter = collections.Counter(flat_list)
    counter= sorted(counter.items(), key=lambda x:x[1],reverse=True)
    sum=0
    temp=[]
    for i in counter:
        if(i[0] not in STOPWORDS and i[0].isalpha() and len(i[0])>2):
            sum+=1
            temp.append(i)
        if(sum==n):
            break
    temp=dict(temp)
    array = [ {'key' : i, 'value' : temp[i]/len(list_of_text)} for i in temp]
    return array
```

```python
import gensim
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem.porter import *
stemmer = PorterStemmer()
from nltk import WordNetLemmatizer
```

```python
def lemmatize_stemming(text):
    return stemmer.stem(WordNetLemmatizer().lemmatize(text, pos='v'))
# Tokenize and lemmatize
def preprocess(text):
    result=[]
    for token in gensim.utils.simple_preprocess(text) :
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(token)
            
    return result
```
